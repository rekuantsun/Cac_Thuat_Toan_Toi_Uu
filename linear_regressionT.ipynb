```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import accuracy_score,r2_score
```

# **2. XÂY DỰNG THUẬT TOÁN**

## **Tập dữ liệu USA_Housing.csv**

### **ĐỌC VÀ CHIA TẬP DỮ LIỆU**

```python
data  = pd.read_csv("/content/drive/MyDrive/Các thuật toán tối ưu/USA_Housing.csv")
del data['Address']

D= np.array(data)
D = preprocessing.MinMaxScaler().fit_transform(D)
X, y = D[:,0], D[:,-1]
X = np.array(X) #đưa X và y về dạng ma trận
y=np.array(y)
X = X.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

### **HỒI QUY ĐƠN BIẾN**

```python
plt.scatter(X, y)
plt.xlabel('trục X')
plt.ylabel('Trục Y')
plt.title('Biểu đồ dữ liệu X và y')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train, y_train)
y_hat = model.predict(X_test)
```

```python
mse=mean_squared_error(y_test, y_hat)
print('Mean Squared Error (MSE):', mse)
rmse = np.sqrt(mean_squared_error(y_test, y_hat))
print('Root Mean Squared Error (RMSE):', rmse)
r2 = r2_score(y_test, y_hat)
print('R-squared (R^2):', r2)
mae = mean_absolute_error(y_test, y_hat)
print("Mean Absolute Error (MAE):", mae)
```

**KẾT LUẬN**

Dựa vào giá trị Mean Squared Error (MSE) của 0.01179041379528232, chúng ta có thể kết luận rằng mô hình đã được đánh giá có hiệu suất tương đối tốt. Khi giá trị MSE gần bằng 0, điều đó cho thấy mô hình dự đoán gần chính xác với dữ liệu thực tế

### **HỒI QUY ĐA BIẾN**
